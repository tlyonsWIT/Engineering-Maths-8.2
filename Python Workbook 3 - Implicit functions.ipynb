## Graphing implicit functions

```
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

### Step 1 - Setting up the (x,y)-plane

```
X=np.linspace(-10,10,501)
Y=np.linspace(-10,10,501)
```

```
x,y=np.meshgrid(X,Y)
```

### Definingin the implicit relation F(x,y)


```
def F(x,y):
    return y+np.sin(x*y)
```

```
plt.contour(x,y, F(x,y),levels=[0])
plt.ylim(-2,2)
plt.savefig('fig1.pdf')
```

```
files.download('fig1.pdf')
```

### Interpreting the implicit function


* We see that $F(x,y)$ is a function defined over the $(x,y)$-plane.

* This means we should input an $x$-value and a $y$-value into this function.

* When we do this, we can interpret the value of the function $F(x,y)$ as the __height__ of a surface above the point $(x,y)$.

* When we use $F(x,y)=0$ to define all $(x,y)$-values such that the function/surface has __height 0__ above the points $(x,y)$ satisfying this equation.

*

```
from matplotlib import cm as cm  ## Library of functions for ColourMaps 
from mpl_toolkits.mplot3d import Axes3D ## Library of functions for 3D plotting
```

```
X=np.linspace(-5,5,501)
Y=np.linspace(-5,5,501)
```

```
x,y=np.meshgrid(X,Y)
```

```
fig = plt.figure(figsize=(12,9))
ax = fig.gca(projection='3d')
ax.plot_surface(x,y,F(x,y),color='cornflowerblue',alpha=0.25) # Plot the surface 

ax.contour(x, y, F(x,y), cmap=cm.bone) # Plot the contours on the surface

#ax.contour(x, y, F(x,y), zdir='z', offset=0, cmap=cm.copper) # Plot the level curves in the (x,y)-plane

ax.set_xlabel('x',size=16) # Set the x-label
ax.set_ylabel('y',size=16) # Set the y-label
ax.set_zlabel('z',size=16) # Set the z-label

ax.set_zlim(-10,10) # set the vertical limits of the plot at -40 (lower) and +20 (upper)
```
